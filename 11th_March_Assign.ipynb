Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Both t-tests and z-tests are statistical tests used to make inferences about population parameters based on sample data, particularly when comparing means. However, they have distinct differences, primarily related to the characteristics of the data and the assumptions that need to be met.

T-Test:
A t-test is used when the sample size is relatively small (typically less than 30) and the population standard deviation is unknown. There are two main types of t-tests: the one-sample t-test, used to compare a sample mean to a known population mean or a hypothesized value, and the two-sample t-test, used to compare the means of two independent samples.

Example Scenario for a T-Test:
Suppose you want to determine if a new teaching method has a significant impact on students' test scores. You randomly select a group of 20 students, teach them using the new method, and then compare their average test scores to a hypothesized population mean of 75. Since the population standard deviation is unknown, you would use a one-sample t-test to determine if the difference in means is statistically significant.

Z-Test:
A z-test is used when the sample size is larger (typically greater than 30) or when the population standard deviation is known. Z-tests are more suitable for cases where you are comparing a sample mean to a known population mean or hypothesized value.

Example Scenario for a Z-Test:
Suppose you are studying the heights of adult males in a certain city. You have access to a dataset of 1000 individuals with their heights, and you want to determine if the average height in this city is significantly different from the national average height of 68 inches. Since you have a large sample size and can reasonably assume the population standard deviation is known, you would use a z-test to compare the sample mean to the hypothesized population mean.

In summary:

(i) Use a t-test when the sample size is small and the population standard deviation is unknown.
(ii) Use a z-test when the sample size is larger (typically > 30) or when the population standard deviation is known.

Q2: Differentiate between one-tailed and two-tailed tests.

One-Tailed Test:
A one-tailed test is a type of hypothesis test where the critical region (the area of the distribution that leads to rejecting the null hypothesis) is located entirely in one tail of the probability distribution. In other words, you are testing for the possibility of a specific directional effect. One-tailed tests are used when you have a specific hypothesis about the direction of the effect you're investigating.

For example, let's say you're testing a new drug's effectiveness in reducing blood pressure. Your null hypothesis (H0) could be that the drug has no effect, and your alternative hypothesis (Ha) could be that the drug reduces blood pressure. If you're only interested in whether the drug reduces blood pressure (and not if it increases it), you would perform a one-tailed test in the direction of reduction.

Two-Tailed Test:
A two-tailed test is another type of hypothesis test where the critical region is split between both tails of the probability distribution. This test is used when you're interested in determining if there's any significant difference between groups or conditions, regardless of the direction of the effect. Two-tailed tests are more conservative because they consider the possibility of an effect in either direction.

Continuing with the drug example, if you want to test whether the drug has any effect on blood pressure (whether it reduces or increases it), you would perform a two-tailed test. Your null hypothesis (H0) would be that the drug has no effect, and your alternative hypothesis (Ha) would be that the drug has a significant effect (either reduction or increase).

In summary:

(i)One-Tailed Test: Used when you have a specific hypothesis about the direction of the effect and are only interested in one tail of the distribution.

(ii)Two-Tailed Test: Used when you want to determine if there is a significant difference between groups or conditions, regardless of the direction of the effect, and the critical region is split between both tails of the distribution.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Type 1 and Type 2 errors are concepts in hypothesis testing that describe the potential mistakes or errors that can occur when making decisions based on statistical tests.

Type 1 Error (False Positive):
A Type 1 error occurs when the null hypothesis (H0) is incorrectly rejected when it is actually true. In other words, you conclude that there is a significant effect or difference when, in reality, there is no such effect or difference in the population.

Example Scenario for Type 1 Error:
Suppose a pharmaceutical company is testing a new drug's effectiveness in reducing cholesterol levels. The null hypothesis (H0) is that the drug has no effect on cholesterol levels. However, due to a random variation in the sample data, the statistical test indicates a significant reduction in cholesterol levels. If the researchers conclude that the drug is effective and reject the null hypothesis, but in reality, the drug has no effect, this would be a Type 1 error.

Type 2 Error (False Negative):
A Type 2 error occurs when the null hypothesis (H0) is incorrectly not rejected when it is actually false. In other words, you fail to detect a significant effect or difference that does exist in the population.

Example Scenario for Type 2 Error:
Continuing with the drug example, let's say the new drug does indeed have a significant effect on reducing cholesterol levels, but the sample size used for testing is too small or the drug's effect is relatively weak. As a result, the statistical test fails to detect the significant reduction in cholesterol levels, and the researchers fail to reject the null hypothesis. This would be a Type 2 error.

In summary:

Type 1 Error (False Positive): Rejecting the null hypothesis when it is actually true.

Example: Incorrectly concluding that a new drug is effective when it has no effect.

Type 2 Error (False Negative): Failing to reject the null hypothesis when it is actually false.

Example: Failing to detect a significant reduction in cholesterol levels when the drug is indeed effective.

Q4: Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new evidence. It provides a way to combine prior knowledge with new information to make more accurate predictions or inferences.

The formula for Bayes's theorem is as follows:

P(A∣B)= P(B∣A)⋅P(A)/ P(B)
                 
Where:

P(A∣B) is the probability of event A occurring given that event B has occurred.

P(B∣A) is the probability of event B occurring given that event A has occurred.

P(A) is the prior probability of event A occurring.

P(B) is the probability of event B occurring.

Now, let's illustrate Bayes's theorem with an example:

Example: Medical Test

Imagine a medical scenario where a patient is being tested for a rare disease. Let's define the following events:

A: The patient has the disease.
B: The test result is positive.
We are given the following probabilities:

P(A): The prior probability that a randomly selected person has the disease. Let's say this is 0.01 (1% of the population).

P(B∣A): The probability that the test result is positive given that the patient has the disease. Let's say this is 0.98 (98% accurate for true positives).

P(B∣¬A): The probability that the test result is positive given that the patient does not have the disease. Let's say there is a 0.02 false positive rate, so P(B∣¬A)=0.02.

We want to find P(A∣B), which is the probability that the patient actually has the disease given that the test result is positive.

Using Bayes's theorem:

P(A∣B)= P(B∣A)⋅P(A)/P(B)

First, we need to calculate P(B):

P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A)

Given that P(¬A)=1−P(A)=0.99, we can calculate P(B) as:

P(B)=(0.98⋅0.01)+(0.02⋅0.99)=0.0196+0.0198=0.0394

Now we can calculate P(A∣B):

P(A∣B)= P(B∣A)⋅P(A)/P(B)
     = 0.98⋅0.01/0.0394
     ≈0.2487
     
So, even though the test is accurate (98% true positive rate), the probability that a patient actually has the disease given a positive test result is only about 25%. This demonstrates the importance of considering both the test accuracy and the prior probability when interpreting test results using Bayes's theorem.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a range of values within which we expect a population parameter, such as a mean or proportion, to fall with a certain level of confidence. It provides a measure of the uncertainty associated with estimating a population parameter based on a sample from that population.

When constructing a confidence interval, we specify a confidence level, typically expressed as a percentage, which represents the likelihood that the true parameter value falls within the interval. For example, a 95% confidence interval implies that if we were to repeat the process of obtaining samples and constructing intervals many times, about 95% of those intervals would contain the true population parameter.

The formula for calculating a confidence interval for the mean of a population (assuming a large enough sample size) is given by:

Confidence Interval = Sample Mean±Margin of Error

Where the margin of error is determined by the standard error of the sample mean and the critical value from the appropriate statistical distribution (usually the t-distribution for small sample sizes or the z-distribution for large sample sizes).

Example: Calculating a Confidence Interval for a Mean

Suppose you are conducting a study to estimate the average weight of newborn babies in a particular hospital. You collect a random sample of 50 newborns and measure their weights. The sample mean weight is 3.2 kilograms, and the sample standard deviation is 0.4 kilograms.

You want to calculate a 95% confidence interval for the true average weight of newborn babies in this hospital.

1.Find the critical value:
For a 95% confidence interval and a sample size of 50, you would use the t-distribution. The degrees of freedom for this distribution would be n−1=49. You can find the critical value from a t-table or calculator. Let's assume the critical value is 2.009 (rounded for simplicity).

2.Calculate the standard error:
Standard Error = Sample Standard Deviation/sqrt(Sample Size)
               = 0.4/sqrt(50)
               ≈ 0.0566

3.Calculate the margin of error:
Margin of Error=Critical Value×Standard Error=2.009×0.0566≈0.114

4.Calculate the confidence interval:
Confidence Interval = Sample Mean ± Margin of Error
                    = 3.2 ± 0.114

Confidence Interval=Sample Mean±Margin of Error=3.2±0.114

The 95% confidence interval for the true average weight of newborn babies in this hospital is approximately 3.086 kg to 3.314 kg. This means that we are 95% confident that the true population average weight falls within this interval.

Remember that the confidence interval provides a range of plausible values for the population parameter based on the sample data and the chosen confidence level. The wider the confidence interval, the less precise our estimate is, but it reflects a higher degree of confidence.

Sample Problem:

You also have information that a special sensor is 95% accurate at detecting defective bulbs (Event B given A) and correctly identifying non-defective bulbs (Event ¬B given ¬A).

Now, you randomly select a light bulb from the production line, and the sensor indicates that it's defective (Event B). What is the probability that the bulb is actually defective (Event A)?

Solution:
Let's define the events:

A: The bulb is defective.

¬A: The bulb is not defective (i.e., it is functional).

B: The sensor indicates that the bulb is defective.

¬B: The sensor indicates that the bulb is not defective.
We are given the following probabilities:

P(A)=0.05: Prior probability of a bulb being defective

P(¬A)=1−P(A)=0.95: Prior probability of a bulb being functional.

P(B∣A)=0.95: Probability of the sensor indicating a bulb is defective when it is actually defective.

P(¬B∣¬A)=0.95: Probability of the sensor indicating a bulb is not defective when it is actually functional.

We want to find P(A∣B), which is the probability that the bulb is defective given that the sensor indicates it's defective.

Using Bayes' Theorem:

P(A∣B)= P(B∣A)⋅P(A)/P(B)

We need to calculate P(B):

P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A)
P(B)=(0.95⋅0.05)+(0.05⋅0.95)=0.0475+0.0475=0.095

Now we can calculate P(A∣B):

P(A∣B)= P(B∣A)⋅P(A)/P(B)
     = 0.95⋅0.05/0.095
     ≈0.5
 
So, the probability that the bulb is actually defective given that the sensor indicates it's defective is approximately 0.5 or 50%.

Even though the sensor is quite accurate (95% true positive rate), due to the relatively low prior probability of a bulb being defective (5%), there is still significant uncertainty in the conclusion based solely on the sensor's indication. This example demonstrates how Bayes' Theorem allows us to update our probabilities based on both prior knowledge and new evidence.

In other words, the margin of error accounts for the variability that exists between different samples that could be drawn from the same population. A larger margin of error indicates a wider confidence interval, which means a less precise estimate of the population parameter, while a smaller margin of error results in a narrower confidence interval and a more precise estimate.

Sample Size and Margin of Error:
Sample size has a significant effect on the margin of error. As the sample size increases, the margin of error decreases, leading to a more precise estimate. This relationship is due to the fact that larger samples provide more information about the population and reduce the variability observed in different samples.
